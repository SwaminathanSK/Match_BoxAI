```python
import numpy as np
import pickle
import random
```

```python
def base_3_to_10(inp):
    input_as_int = 0
    for i in range(9, 0, -1):
        input_as_int += int(inp[9-i])*(3**(i-1))
    return input_as_int

def reinforce(history, verdict, old_li):
    new_li = old_li
    if verdict == 1: # x wins
        for i in history:
            new_li[i[0]][1][i[1]] += 2
    elif verdict == 2: # x loses
        for i in history:
            new_li[i[0]][1][i[1]] -= 1
    elif verdict == 3: # draw
        for i in history:
            new_li[i[0]][1][i[1]] += 1
    return new_li

def verdict_check(n, move, li): # x win check = 1, o win check = 2
    li2 = [int(i) for i in li[n][0]]
    verdict = -1 # game continues
    for x_0 in [1, 2]:
        verdict_check = ((li2[0]==x_0)+(li2[1]==x_0)+(li2[2]==x_0) == 3) +((li2[3]==x_0)+(li2[4]==x_0)+(li2[5]==x_0) == 3) +((li2[6]==x_0)+(li2[7]==x_0)+(li2[8]==x_0) == 3) +((li2[0]==x_0)+(li2[3]==x_0)+(li2[6]==x_0) == 3) +((li2[1]==x_0)+(li2[4]==x_0)+(li2[7]==x_0) == 3) +((li2[2]==x_0)+(li2[5]==x_0)+(li2[8]==x_0) == 3) +((li2[0]==x_0)+(li2[4]==x_0)+(li2[8]==x_0) == 3) +((li2[2]==x_0)+(li2[4]==x_0)+(li2[6]==x_0) == 3)
        if verdict_check > 0:
            verdict = x_0
            break
    if verdict == -1 and move.count('0') == 0: # draw condition
        verdict = 3
    return verdict

def move(n, history, li): # n is the (index+1) of the mapping to the board config
    li2 = li[n]
    verdict = verdict_check(n, li2[0], li)
    li = reinforce(history, verdict, li)
    move = li2[0]
    
    if verdict != 2:
        draw = random.choices([0, 1, 2, 3, 4, 5, 6, 7, 8], weights = li2[1], k = 1)
        draw = draw[0]
        move = li2[0]
        move = move[:draw] + '1' + move[draw+1:]


        history.append((n, draw))

        verdict = verdict_check(base_3_to_10(move), move, li)
        li = reinforce(history, verdict, li)               
    return (verdict, history, li, move)

                            
with open('Boards.bin', 'rb') as file:
    li = pickle.load(file)
                            
# history is a list of tuples -> (previous move index, choice of index from the appendable)
history = [] # add this to a function later
verdict = -1

(verdict, history, li, result) = move(0, history, li)
print()
print(result[:3])
print(result[3:6])
print(result[6:])
print()

while(verdict == -1):
    row_1 = input()
    row_2 = input()
    row_3 = input()
    inp = row_1 + row_2 + row_3
    input_as_int = base_3_to_10(inp)
    (verdict, history, li, result) = move(input_as_int, history, li)
    if verdict != 2:
        print()
        print(result[:3])
        print(result[3:6])
        print(result[6:])
        print()
    else:
        print("Bruh u win")

with open('Boards.bin', 'wb') as file:
    pickle.dump(li, file)
                     
```
